### Fill VectorStore

```python
from openai import OpenAI
from intergrax.openai.rag.rag_openai import IntergraxRagOpenAI
from dotenv import load_dotenv
import os
load_dotenv()

os.environ['OPENAI_API_KEY'] = os.getenv('OPENAI_API_KEY')
VECTOR_STORE_ID = os.getenv("VECTOR_STORE_ID")

# use your folder with data (pdf, txt, docx)
LOCAL_FOLDER = "../documents/mooff-strategy/txt"

client = OpenAI()

rag = IntergraxRagOpenAI(client=client, vector_store_id=VECTOR_STORE_ID)
rag.ensure_vector_store_exists()
rag.clear_vector_store_and_storage()
rag.upload_folder_to_vector_store(LOCAL_FOLDER)
```
